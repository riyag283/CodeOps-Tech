```python
import os
import json
import pandas as pd

filepath = os.path.join(os.getcwd(),"filejson1.json")
file = open(filepath)
data = json.loads(file.read())
file.close()
```

```python
pdict = {}
k = len(data['paid'])
h = len(data['free'])
for i in range(k):
    pdict[i] = data['paid'][i]
pdf = pd.DataFrame(pdict, columns=pdict.keys())
pdf = pdf.T

fdict = {}
for i in range(h):
    fdict[i+k] = data['free'][i]
fdf = pd.DataFrame(fdict, columns=fdict.keys())
fdf = fdf.T
```

```python
df = pd.concat([pdf,fdf])
oridata = df
```

```python
def isInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
```

```python
def hasNumbers(inputString):   
    return any(char.isdigit() for char in inputString)
```

```python
def check_match(sen1,sen2):
    small = len(sen1) < len(sen2) and sen1 or sen2
    big = len(sen1) >= len(sen2) and sen1 or sen2
    list1 = big.split(' ')
    flag = 0
    index = 0
    for word in list1:
        if isInt(word) and len(word)==4:
            word = word[2:]
            if hasNumbers(small):
                if word not in small:
                    return False
            else:
                continue
        if word[:1] not in small:
            percentmatch = (index/len(small))*100
            if percentmatch >= 80:
                continue
            else:
                return False
        index = small.find(word[:2])
        small = small[index:]
    return True
```
